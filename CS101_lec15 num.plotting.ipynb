## Basic Plots

```python
from matplotlib.pyplot import plot,show
```

```python
import numpy as np
x = np.linspace( 0,1,11 )
y = np.sin( np.pi * x )
```

```python
plot( x,y )
show()
```

## Data Plots (two-dimensional)

Data can be two-dimensional, in which case line plots may not be the best way to display the data. For these cases, use imshow:

```python
from matplotlib.pyplot import imshow,show
```

```python
data = np.loadtxt( 'inflammation15.csv',delimiter=',' )

imshow( data )
show()
```

## Importing MatPlotLib/PyPlot

```python
import matplotlib.pyplot as plt
```

```python
x = [ 0,1,2,3,4 ]
y = [ 0,1,2,1,0 ]
plt.plot( x,y,'ro' )
plt.show()
```

```python
from math import pi
x = []   # can't use range easily!
for i in range(100):
    x.append( 2*pi*i/100 )
from math import sin
y = []
for j in range(100):
    y.append( sin(x[j]) )

import matplotlib.pyplot as plt
plt.plot( x,y,'k-' ,c='red')
plt.xlim( 0,2*pi )
plt.ylim( -1,1 )
plt.title("123")
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['red'])#
plt.show()
```

Use numpy is the same.

```python
import numpy as np

x = np.linspace( 0,2*np.pi,101 )
y = np.sin( x )

import matplotlib.pyplot as plt
plt.plot( x,y,'k-' )
plt.xlim( 0,2*np.pi )
plt.ylim( -1,1 )
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace( -np.pi,np.pi,101 )
x = np.cos( t )
y = np.sin( t )

plt.plot( x,y )
plt.xlabel( 'x Axis' )
plt.ylabel( 'y Axis' )
plt.title( 'Line Graph')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace( 0,2*np.pi,51)
plt.plot( x,np.sin(x),'b-')
plt.plot( x,np.sin(x)**2,'b--')
plt.plot( x,np.tan(x),'r-')
plt.plot( x,np.sinh(x)/250,'g-')
plt.xlim( 2,4 )
plt.ylim( -2,2 )
plt.title('Trigonometry Functions')
plt.show()
```
