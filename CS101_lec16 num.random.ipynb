## Distributions

### Uniform Distribution 

```python
import numpy as np
numbers = np.random.uniform( size=(1000,) )

import matplotlib.pyplot as plt
plt.plot( numbers,'bo' )
plt.show()
```

```python
chance = np.random.uniform()
if chance < 0.30:
    print( 'it happened!' )
```

### Normal Distribution

```python
numbers = np.random.normal( size=(1000,) )

import matplotlib.pyplot as plt
plt.plot( numbers,'bo' )
plt.show()
```

```python
mean = 10.0
stdev = 2.0
np.random.normal( size=(100 ) ) * stdev + mean
```

### Discrete Distributions （randint）

```python
numbers = np.random.randint( 10,size=(1000,) )
# 无法取到10
import matplotlib.pyplot as plt
plt.plot( numbers,'r.' )
plt.show()
```

### randomized (shuffled) set

```python
deck = list( range( 52 ) )
np.random.shuffle( deck )
#将deck重新排序
```

### random selection

```python
deck = list( range( 52 ) )
np.random.choice( deck )
#随机抽取
```

## Histograms

```python
import numpy as np
numbers = np.random.normal( size=(1000,) )
import matplotlib.pyplot as plt
plt.hist( numbers,bins=10 )
plt.show()
```
